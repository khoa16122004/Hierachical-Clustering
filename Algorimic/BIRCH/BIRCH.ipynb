```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title('Dữ liệu hình trăng lưỡi liềm')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

```

```python
from sklearn.cluster import KMeans, Birch
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

birch = Birch(n_clusters=2)
birch_labels = birch.fit_predict(X)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o', edgecolor='k')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=birch_labels, cmap='viridis', marker='o', edgecolor='k')
plt.title('BIRCH Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('t4.8k.csv', header=None, delim_whitespace=True)
print(data.head())
print(data.info())
X = data.values
print(X.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, Birch

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

birch = Birch(threshold=0.1, branching_factor=50, n_clusters=5)
birch_labels = birch.fit_predict(X)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o', edgecolor='k')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=birch_labels, cmap='viridis', marker='o', edgecolor='k')
plt.title('BIRCH Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
