## Smiling Face

```python
# Load data
file_path = '../2D_Data/smileface.csv'
data = pd.read_csv(file_path)
data.columns = ['X', 'Y']
X_smiling = data.values
```

```python
# Library
clustering = InteractiveHierarchicalClustering2D(data=X_smiling, max_clusters=10, linkage_method='ward', p=5)
clustering.plot_dendrogram()
interact(clustering.interactive_clustering, n_clusters=IntSlider(min=1, max=clustering.max_clusters, step=1, value=1))
clustering.save_clustering_animation(save_path='./Visualization/Lib_SmilingFace.gif')
```

```python
# BisectingKMeans
clustering = BisectingKmeansClustering(X_smiling, max_clusters=5)
clustering.fit()
clustering.interactive()
clustering.create_gif("./Visualization/BisectingKmeansClustering_SmilingFace.gif")
```

```python
# MST
clustering = MSTDivisiveClustering(X_smiling, max_clusters=6, max_cuts=6)
clustering.divisive_clustering_mst_interactive()
clustering.create_gif("./Visualization/mst_Smiling.gif")
```
